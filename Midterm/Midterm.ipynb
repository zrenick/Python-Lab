```python
import matplotlib.pyplot as plt
# Create an empty array to store the 5000 samples of size 1000 for both variables
x_samples = np.empty((5000, 1000))
y_samples = np.empty((5000, 1000))

# Now sample with replacement 5000 times and store the samples in the empty arrays using numpys np.random.choice function
for i in range(5000):
    sample_indices = np.random.choice(range(100), size=1000, replace=True)
    x_samples[i] = x[sample_indices]
    y_samples[i] = y[sample_indices]

#Calculate the median value of x and y (Using the axis option 'none' to perform median on the entire flattened array to get the single median)
x_median = np.median(x_samples, axis=None)
y_median = np.median(y_samples, axis=None)
x_median
```

```python
# Create new empty arrays to store the 5000 samples of size 1000 for both variables
x_2_samples = np.empty((5000, 1000))
y_2_samples = np.empty((5000, 1000))

# Now sample with replacement the same way we did before
for i in range(5000):
    sample_indices = np.random.choice(range(100), size=1000, replace=True)
    x_2_samples[i] = x[sample_indices]
    y_2_samples[i] = y[sample_indices]

def IQR_SD(sample):
    # calculate Q1, Q3, and IQR
    q1 = np.percentile(sample, 25)
    q3 = np.percentile(sample, 75)
    iqr = q3 - q1
    std = np.std(sample)
    return iqr, std

```

```python
# I realized after doing the previous part that it makes more sense to have a 2D numpy array 
# where the first column contains the x values and the second column contains the y values 
# and then pass that into the function so I could calculate the IQR and std for both variables at once.
# However this meant I needed to change how I did the random sampling of the arrays. 
# So I instead made one empty 2D array for storage and then sampled the x and y arrays at the same time and stacked them using numpys vp.stack function .
# But this made the 2D array have the x and y arrays as each row which was weird to pass into the function 
# and didn't lead to a nice looking result dataframe, so I transposed it instead. 

# Create 5000 samples with replacement like before but combining x and y into one 2D array
samples = np.empty((5000, 1000, 2))
for i in range(5000):
    indices = np.random.choice(range(100), size=1000, replace=True)
    samples[i] = np.vstack((x[indices], y[indices])).T #Transpose to make the variables as the columns

# Function to calculate IQR and SD of x and y
def calculate_stats(sample):
    iqr_x = np.percentile(sample[:, 0], 75) - np.percentile(sample[:, 0], 25) #take all rows and column 0 "x"
    iqr_y = np.percentile(sample[:, 1], 75) - np.percentile(sample[:, 1], 25) #take all rows and column 1 "y"
    sd_x = np.std(sample[:, 0])
    sd_y = np.std(sample[:, 1])
    return iqr_x, iqr_y, sd_x, sd_y

# Calculate IQR and SD for each sample
results = np.empty((5000, 4))
for i in range(5000):
    results[i] = calculate_stats(samples[i])

# Summary of IQR and SD values
df_results = pd.DataFrame(results, columns=['IQR_x', 'IQR_y', 'SD_x', 'SD_y'])
df_results.head(10)
```

```python
import numpy as np
# Generate a random sample of size 10,000 from the standard normal distribution
np.random.seed(123)
original_sample = np.random.normal(0, 1, size=10000)
```

i.

```python
sample_median = np.median(original_sample)
sample_mean = np.mean(original_sample)
print('Sample mean:', sample_mean)
print('Sample median:', sample_median)
```

ii.

```python
import numpy as np
import pandas as pd

# Setting 50,000 bootsrap samples
num_bootstrap = 50000

# Making an empty data frame to store the bootstrap sample mean and median
bootstrap_results = pd.DataFrame(columns=['Mean', 'Median'])

# Generate bootstrap samples and calculate mean and median for each
for i in range(num_bootstrap):
    bootstrap_sample = np.random.choice(original_sample, size=10000, replace=True) #using values from the original sample to create a new "fake" dataset
    bootstrap_mean = np.mean(bootstrap_sample) #taking the mean of this new "fake" dataset
    bootstrap_median = np.median(bootstrap_sample) #taking the median of this new "fake" dataset
    bootstrap_results.loc[i] = [bootstrap_mean, bootstrap_median] #storing the mean and median in our empty datafram for the sample interation out of 50,000 we are on. 

# Now we have a dataframe of a bunch of differnt bootstrapped means and medians 
# We can take the mean and median respectively of these columns in the dataframe to get an estimate of the true population mean and median.
bootstrap_mean_estimate = np.mean(bootstrap_results['Mean'])
bootstrap_median_estimate = np.median(bootstrap_results['Median'])

# Store and Display the results. 
boot_mean = bootstrap_mean_estimate
boot_median = bootstrap_median_estimate
print('Bootstrap estimate of mean:', boot_mean)
print('Bootstrap estimate of median:', boot_median)
```

iii.

```python
# Lets reuse the original random sample from the previous part to be our starting dataset

# Define a function to be applied to each subsample to calculate the mean and median of the original sample data
def statistic(data):
    return np.mean(data) , np.median(data)

# Calculate the mean and median for the original sample
sample_statistic = statistic(original_sample)

# Create jackknife resamples
n = len(original_sample)
jackknife_samples = np.empty((n, n-1)) #an empty array of size 10000 to store our results of the jacknife samples
# each row corresponds to one of the jackknife samples and each column corresponds to an observation that is omitted in that particular jackknife sample.
# The column is n-1 becasue we start with only 1 observation left out, but as we iterate we leave more and more out until we leave out 9999 observations.
for i in range(n):
    jackknife_samples[i] = np.concatenate((original_sample[:i], original_sample[i+1:])) #use concatenate to create a new sample dataset (a new array with one element removed)
    # We are selecting up to i indexs of the original dataset and then skipping one observation index to leave it out and add the rest of the data after it.

# Calculate the mean and median for each jackknife resample using the created function and store them in an array
# This array will have two values mean and median for each observation
jackknife_statistics = np.array([statistic(sample) for sample in jackknife_samples])

#Calculating the grand mean and median over the entire array of the jackknifed samples
jack_mean = np.mean(jackknife_statistics)
jack_median = np.median(jackknife_statistics)

# Display the results. 
print('Jackknife estimate of mean:', jack_mean)
print('Jackknife estimate of median:', jack_median)
```

iv.

```python
import numpy as np
from scipy.stats import bootstrap

# Convert the original sample to a tuple so it can be properly passed into the bootstrap function
sample = tuple(original_sample)
t = (sample,)

# Use bootstrap function to estimate sample mean and median
mean_estimate = bootstrap(t, statistic=np.mean, n_resamples=50000)
median_estimate = bootstrap(t, statistic=np.median, n_resamples=50000)

# Store and Print the results
scipy_mean = mean_estimate.bootstrap_distribution.mean() #go into the bootstrap distribution part of the BootstrapResult to grab all of the mean values we calculated and take the mean of all of those. 
scipy_median = median_estimate.bootstrap_distribution.mean() #similarly for the medians. 
print("Bootstrap mean estimate:",  scipy_mean)
print("Bootstrap median estimate:",  scipy_median)
```

v.

```python
# create a list of tuples containing the name of the bootstrapping technique and the mean and median obtained
results = [('Sample', sample_mean, sample_median),
           ('Manual Bootstrapping', boot_mean, boot_median),
           ('Jackknife Resampling', jack_mean, jack_median),
           ('Scipy Bootstrap', scipy_mean, scipy_median)]

# create a dataframe from the list of tuples for each of the sampling techniques
df = pd.DataFrame(results, columns=['Technique', 'Mean', 'Median'])

# display the dataframe
print(df.to_string(index=False))

```
