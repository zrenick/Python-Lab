```python
# Part 2

def player_connect(player):
    temp = d[d.PLAYER_NAME == player]
    temp = pd.DataFrame(temp)
    year = temp['SEASON'].unique()
    players = []
    
    if len(year) > 0:
        for y in year:
            temp2 = d[d.SEASON == y]
            players.append(temp2.PLAYER_NAME.unique())
    else:
        players = []
        print('Sorry no player found!!!')

    return(players)
```

```python
player1 = 'Bojan Bogdanovic'
player_connect(player1)
```
